```python
def add_ubigeo_fields_to_CEM_df(ubigeo,cem_df):
    DEP = []
    PROV = []
    DISTR = []
    for i in range(len(cem_df['CEM'])):
        for j in range(len(ubigeo['CEM'])):
            if(cem_df['CEM'][i] == ubigeo['CEM'][j]):
                DEP.append(ubigeo['DEP'][j])
                PROV.append(ubigeo['PROV'][j])
                DISTR.append(ubigeo['DISTR'][j])
                break
    cem_df = cem_df.assign(DEP = DEP)
    cem_df = cem_df.assign(PROV = PROV)
    cem_df = cem_df.assign(DISTR = DISTR)
    return cem_df
```

```python
CEM_df_2017_with_ubigeo = add_ubigeo_fields_to_CEM_df(UBIGEO,CEM_df_2017)
CEM_df_2018_with_ubigeo = add_ubigeo_fields_to_CEM_df(UBIGEO,CEM_df_2018)
CEM_df_2019_with_ubigeo = add_ubigeo_fields_to_CEM_df(UBIGEO,CEM_df_2019)
CEM_df_2020_with_ubigeo = add_ubigeo_fields_to_CEM_df(UBIGEO,CEM_df_2020)
```

```python
def generate_days_ubigeo_base(cem_df):
    return cem_df[['DEP','PROV','DISTR','FECHA_INGRESO']]
```

```python
#DEFINITIVA

def poblate_df(days_ubigeo_base,cem_df,year):
    man_physical_v = []
    man_sexual_v = []
    man_economical_v = []
    man_psychological_v = []
    man_first_time = []
    man_inf = []
    man_nin = []
    man_ado = []
    man_ado_t = []
    man_jov = []
    man_adu = []
    man_m60 = []
    man_family = []
    man_love = []
    man_no_rel = []
    man_aggr_alcohol = []
    man_aggr_drug =[]
    man_vict_disab = []
    man_vict_alco = []
    man_vict_drug = []
    man_vict_lgtbi = []
    
    sex = []
    for j in range(len(cem_df['DISTR'])):
        man_physical_v_count = 0
        man_sexual_v_count = 0
        man_economical_v_count = 0
        man_psychological_v_count = 0
        man_first_time_count = 0
        man_inf_count = 0
        man_nin_count = 0
        man_ado_count = 0
        man_ado_t_count = 0
        man_jov_count = 0
        man_adu_count = 0
        man_m60_count = 0
        man_family_count = 0
        man_love_count = 0
        man_no_rel_count = 0
        man_aggr_alcohol_count = 0
        man_aggr_drug_count =0
        man_vict_disab_count = 0
        man_vict_alco_count = 0
        man_vict_drug_count = 0
        man_vict_lgtbi_count = 0

        sex.append(cem_df['SEXO_VICTIMA'][j])

        #edad
        if(cem_df['EDAD_VICTIMA'][j]<6):
            man_inf_count+=1
        elif(cem_df['EDAD_VICTIMA'][j]<12):
            man_nin_count+=1
        elif(cem_df['EDAD_VICTIMA'][j]<15):
            man_ado_count+=1
        elif(cem_df['EDAD_VICTIMA'][j]<18):
            man_ado_t_count+=1
        elif(cem_df['EDAD_VICTIMA'][j]<30):
            man_jov_count+=1
        elif(cem_df['EDAD_VICTIMA'][j]<60):
            man_adu_count+=1
        else:
            man_m60_count+=1

        #tipo_violencia
        if(cem_df['TIPO_VIOLENCIA'][j]==0):
            man_economical_v_count+=1
        elif(cem_df['TIPO_VIOLENCIA'][j]==1):
            man_psychological_v_count+=1
        elif(cem_df['TIPO_VIOLENCIA'][j]==2):
            man_physical_v_count+=1
        elif(cem_df['TIPO_VIOLENCIA'][j]==3):
            man_sexual_v_count+=1

        #primera_vez
        if(cem_df['PRIMERA_VEZ_AGREDE'][j] == 1):
            man_first_time_count+=1

        #factores
        if(cem_df['FACTOR_AGRESOR_CONSUMO_ALCOHOL'][j] == '1'):
            man_aggr_alcohol_count+=1
        elif(cem_df['FACTOR_AGRESOR_CONSUME_DROGA'][j] == '1'):
            man_aggr_drug_count+=1
        elif(cem_df['FACTOR_VICTIMA_DISCAPACIDAD'][j] == '1'):
            man_vict_disab_count+=1
        elif(cem_df['FACTOR_VICTIMA_ABUSO_CONSUMO_ALCOHOL'][j] == '1'):
            man_vict_alco_count+=1
        elif(cem_df['FACTOR_VICTIMA_CONSUME_DROGAS'][j] == '1'):
            man_vict_drug_count+=1
        elif(year in [2019,2020] and cem_df['VULNERABILIDAD_VICTIMA_LGTBI'][j] == '1'):
            man_vict_lgtbi_count+=1
        elif(year in [2017,2018] and cem_df['FACTOR_VICTIMA_POBLACION_LGTBI'][j] == '1'):
            man_vict_lgtbi_count+=1

        #vinculo
        if(cem_df['VINCULO_PAREJA'][j] != ' '):
            man_love_count+=1
        elif(cem_df['VINCULO_FAMILIAR'][j] != ' '):
            man_family_count+=1
        elif(cem_df['SIN_VINCULO'][j] != ' '):
            man_no_rel_count+=1
                
        man_physical_v.append(man_physical_v_count)
        man_sexual_v.append(man_sexual_v_count)
        man_economical_v.append(man_economical_v_count)
        man_psychological_v.append(man_psychological_v_count)
        man_first_time.append(man_first_time_count)
        man_inf.append(man_inf_count)
        man_nin.append(man_nin_count)
        man_ado.append(man_ado_count)
        man_ado_t.append(man_ado_t_count)
        man_jov.append(man_jov_count)
        man_adu.append(man_adu_count)
        man_m60.append(man_m60_count)
        man_family.append(man_family_count)
        man_love.append(man_love_count)
        man_no_rel.append(man_no_rel_count)
        man_aggr_alcohol.append(man_aggr_alcohol_count)
        man_aggr_drug.append(man_aggr_drug_count)
        man_vict_disab.append(man_vict_disab_count)
        man_vict_alco.append(man_vict_alco_count)
        man_vict_drug.append(man_vict_drug_count)
        man_vict_lgtbi.append(man_vict_lgtbi_count)

    
    #añadimos las nuevas columnas al df
    days_ubigeo_base = days_ubigeo_base.assign(man_physical_v = man_physical_v)
    days_ubigeo_base = days_ubigeo_base.assign(man_sexual_v = man_sexual_v)
    days_ubigeo_base = days_ubigeo_base.assign(man_economical_v = man_economical_v)
    days_ubigeo_base = days_ubigeo_base.assign(man_psychological_v = man_psychological_v)
    
    days_ubigeo_base = days_ubigeo_base.assign(man_first_time = man_first_time)
    
    days_ubigeo_base = days_ubigeo_base.assign(man_inf = man_inf)
    days_ubigeo_base = days_ubigeo_base.assign(man_nin = man_nin)
    days_ubigeo_base = days_ubigeo_base.assign(man_ado = man_ado)
    days_ubigeo_base = days_ubigeo_base.assign(man_ado_t = man_ado_t)
    days_ubigeo_base = days_ubigeo_base.assign(man_jov = man_jov)
    days_ubigeo_base = days_ubigeo_base.assign(man_adu = man_adu)
    days_ubigeo_base = days_ubigeo_base.assign(man_m60 = man_m60)
    
    days_ubigeo_base = days_ubigeo_base.assign(man_family = man_family)
    days_ubigeo_base = days_ubigeo_base.assign(man_love = man_love)
    days_ubigeo_base = days_ubigeo_base.assign(man_no_rel = man_no_rel)
    
    days_ubigeo_base = days_ubigeo_base.assign(man_aggr_alcohol = man_aggr_alcohol)
    days_ubigeo_base = days_ubigeo_base.assign(man_aggr_drug = man_aggr_drug)
    days_ubigeo_base = days_ubigeo_base.assign(man_vict_disab = man_vict_disab)
    days_ubigeo_base = days_ubigeo_base.assign(man_vict_alco = man_vict_alco)
    days_ubigeo_base = days_ubigeo_base.assign(man_vict_drug = man_vict_drug)
    days_ubigeo_base = days_ubigeo_base.assign(man_vict_lgtbi = man_vict_lgtbi)
    
    days_ubigeo_base = days_ubigeo_base.assign(sex = sex)
    
    
    return days_ubigeo_base
```

```python
def add_week(days_ubigeo_base,week_df):
    WEEK = []
    for i in range(len(days_ubigeo_base['DISTR'])):
        for j in range(len(week_df['DIA'])):
            if(days_ubigeo_base['FECHA_INGRESO'][i] == week_df['DIA'][j]):
                WEEK.append(week_df['SEMANA'][j])
                break
    days_ubigeo_base = days_ubigeo_base.assign(WEEK = WEEK)
    return days_ubigeo_base
```
