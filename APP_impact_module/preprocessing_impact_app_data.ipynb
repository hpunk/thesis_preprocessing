```python
import pandas as pd

APP_df_2017 = pd.read_csv("../../data/Data/2017/app_2017_value.csv",delimiter=";")
APP_df_2018 = pd.read_csv("../../data/Data/2018/app_2018_value.csv",delimiter=",")
APP_df_2019 = pd.read_csv("../../data/Data/2019/app_2019_value.csv",delimiter=";")
APP_df_2020 = pd.read_csv("../../data/Data/2020/app_2020_value.csv",delimiter=";")
```

```python
## Saving the columns for examination
APP_2017_columns = APP_df_2017.columns
APP_2018_columns = APP_df_2018.columns
APP_2019_columns = APP_df_2019.columns
APP_2020_columns = APP_df_2020.columns

print("Los datos de APP 2017 tienen",APP_2017_columns.size,"columnas")
print("Los datos de APP 2018 tienen",APP_2018_columns.size,"columnas")
print("Los datos de APP 2019 tienen",APP_2019_columns.size,"columnas")
print("Los datos de APP 2020 tienen",APP_2020_columns.size,"columnas")
```

```python
all_columns = []
for column in APP_2017_columns:
    all_columns.append(column)
for column in APP_2018_columns:
    all_columns.append(column)
for column in APP_2019_columns:
    all_columns.append(column)
for column in APP_2020_columns:
    all_columns.append(column)
    
unique_columns = list(set(all_columns))
print("Columnas únicas en total",len(unique_columns))
```

```python
# Columns from all_columns that are not in 2017
not_in_2017 = []
for column in unique_columns:
    if (column not in APP_2017_columns):
        not_in_2017.append(column)
print("Columnas que no están en 2017 del total:",len(not_in_2017))

# Columns from all_columns that are not in 2018
not_in_2018 = []
for column in unique_columns:
    if (column not in APP_2018_columns):
        not_in_2018.append(column)
print("Columnas que no están en 2018 del total:",len(not_in_2018))

# Columns from all_columns that are not in 2019
not_in_2019 = []
for column in unique_columns:
    if (column not in APP_2019_columns):
        not_in_2019.append(column)
print("Columnas que no están en 2019 del total:",len(not_in_2019))

# Columns from all_columns that are not in 2020
not_in_2020 = []
for column in unique_columns:
    if (column not in APP_2020_columns):
        not_in_2020.append(column)
print("Columnas que no están en 2019 del total:",len(not_in_2020))
```

```python
# Columns in common between data from 2017, 2018, 2019
in_all_years = []
for column in unique_columns:
    if(column in APP_2017_columns):
        if(column in APP_2018_columns):
            if(column in APP_2019_columns):
                if(column in APP_2020_columns):
                    in_all_years.append(column)
print("Columnas presentes en los 4 años:",len(in_all_years))
```
