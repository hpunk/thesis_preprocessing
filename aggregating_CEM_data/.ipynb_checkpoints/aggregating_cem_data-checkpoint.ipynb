```python
def aggregate_for_day(base_df):
    print("comenzando")
    districts = base_df['DISTR'].unique()
    
    fecha =[]
    dep = []
    distr = []
    prov = []
    week = []
    
    physical_v = []
    sexual_v = []
    economical_v = []
    psychological_v = []
    first_time = []
    inf = []
    nin = []
    ado = []
    ado_t = []
    jov = []
    adu = []
    m60 = []
    family = []
    love = []
    no_rel = []
    aggr_alcohol = []
    aggr_drug =[]
    vict_disab = []
    vict_alco = []
    vict_drug = []
    vict_lgtbi = []
    
    for d in districts:
        
        partial_df = base_df[base_df['DISTR'] == d].reset_index(drop=True)
        dates = partial_df['FECHA_INGRESO'].unique()
        
        for j in range(len(dates)):
            physical_v_count = 0
            sexual_v_count = 0
            economical_v_count = 0
            psychological_v_count = 0
            first_time_count = 0
            inf_count = 0
            nin_count = 0
            ado_count = 0
            ado_t_count = 0
            jov_count = 0
            adu_count = 0
            m60_count = 0
            family_count = 0
            love_count = 0
            no_rel_count = 0
            aggr_alcohol_count = 0
            aggr_drug_count =0
            vict_disab_count = 0
            vict_alco_count = 0
            vict_drug_count = 0
            vict_lgtbi_count = 0
            
            first = True
            for i in range(len(partial_df['DISTR'])):
                if(partial_df['FECHA_INGRESO'][i] == dates[j]):
                    if(first):
                        dep.append(partial_df['DEP'][i])
                        distr.append(partial_df['DISTR'][i])
                        fecha.append(dates[j])
                        prov.append(partial_df['PROV'][i])
                        week.append(partial_df['WEEK'][i])
                        first = False
                        
                    physical_v_count += partial_df['man_physical_v'][i]
                    sexual_v_count += partial_df['man_sexual_v'][i]
                    economical_v_count += partial_df['man_economical_v'][i]
                    psychological_v_count += partial_df['man_psychological_v'][i]
                    first_time_count += partial_df['man_first_time'][i]
                    inf_count += partial_df['man_inf'][i]
                    nin_count += partial_df['man_nin'][i]
                    ado_count += partial_df['man_ado'][i]
                    ado_t_count += partial_df['man_ado_t'][i]
                    jov_count += partial_df['man_jov'][i]
                    adu_count += partial_df['man_adu'][i]
                    m60_count += partial_df['man_m60'][i]
                    family_count += partial_df['man_family'][i]
                    love_count += partial_df['man_love'][i]
                    no_rel_count += partial_df['man_no_rel'][i]
                    aggr_alcohol_count += partial_df['man_aggr_alcohol'][i]
                    aggr_drug_count += partial_df['man_aggr_drug'][i]
                    vict_disab_count += partial_df['man_vict_disab'][i]
                    vict_alco_count += partial_df['man_vict_alco'][i]
                    vict_drug_count += partial_df['man_vict_drug'][i]
                    vict_lgtbi_count += partial_df['man_vict_lgtbi'][i]

            physical_v.append(physical_v_count)
            sexual_v.append(sexual_v_count)
            economical_v.append(economical_v_count)
            psychological_v.append(psychological_v_count)
            first_time.append(first_time_count)
            inf.append(inf_count)
            nin.append(nin_count)
            ado.append(ado_count)
            ado_t.append(ado_t_count)
            jov.append(jov_count)
            adu.append(adu_count)
            m60.append(m60_count)
            family.append(family_count)
            love.append(love_count)
            no_rel.append(no_rel_count)
            aggr_alcohol.append(aggr_alcohol_count)
            aggr_drug.append(aggr_drug_count)
            vict_disab.append(vict_disab_count)
            vict_alco.append(vict_alco_count)
            vict_drug.append(vict_drug_count)
            vict_lgtbi.append(vict_lgtbi_count)
            
    #armamos df y retornamos
    data = {'fecha':fecha,'distr':distr, 'prov' : prov, 'dep':dep, 'week':week, 'physical_v':physical_v,
            'sexual_v':sexual_v, 'economical_v':economical_v, 'psychological_v':psychological_v, 'first_time':first_time,
            'inf':inf, 'nin':nin, 'ado':ado,'ado_t':ado_t,'jov':jov,'adu':adu,'m60':m60,'family':family,'love':love,
            'no_rel':no_rel,'aggr_alcohol':aggr_alcohol,'aggr_drug':aggr_drug,'vict_disab':vict_disab,'vict_alco':vict_alco,
            'vict_drug':vict_drug,'vict_lgtbi':vict_lgtbi}
    response_df = pd.DataFrame(data)
    print("acabado")
    return response_df
```

```python
CEM_df_2017_H_DAYS = aggregate_for_day(CEM_df_2017_H)
CEM_df_2017_M_DAYS = aggregate_for_day(CEM_df_2017_M)
CEM_df_2018_H_DAYS = aggregate_for_day(CEM_df_2018_H)
CEM_df_2018_M_DAYS = aggregate_for_day(CEM_df_2018_M)
CEM_df_2019_H_DAYS = aggregate_for_day(CEM_df_2019_H)
CEM_df_2019_M_DAYS = aggregate_for_day(CEM_df_2019_M)
CEM_df_2020_H_DAYS = aggregate_for_day(CEM_df_2020_H)
CEM_df_2020_M_DAYS = aggregate_for_day(CEM_df_2020_M)
```

```python
#AGREGAMOS POR SEMANA
def aggregate_for_week(base_df):
    print("comenzando")
    districts = base_df['distr'].unique()
    
    fecha =[]
    dep = []
    distr = []
    prov = []
    week = []
    
    physical_v = []
    sexual_v = []
    economical_v = []
    psychological_v = []
    first_time = []
    inf = []
    nin = []
    ado = []
    ado_t = []
    jov = []
    adu = []
    m60 = []
    family = []
    love = []
    no_rel = []
    aggr_alcohol = []
    aggr_drug =[]
    vict_disab = []
    vict_alco = []
    vict_drug = []
    vict_lgtbi = []
    
    for d in districts:
        
        partial_df = base_df[base_df['distr'] == d].reset_index(drop=True)
        dates = partial_df['week'].unique()
        
        for j in range(len(dates)):
            physical_v_count = 0
            sexual_v_count = 0
            economical_v_count = 0
            psychological_v_count = 0
            first_time_count = 0
            inf_count = 0
            nin_count = 0
            ado_count = 0
            ado_t_count = 0
            jov_count = 0
            adu_count = 0
            m60_count = 0
            family_count = 0
            love_count = 0
            no_rel_count = 0
            aggr_alcohol_count = 0
            aggr_drug_count =0
            vict_disab_count = 0
            vict_alco_count = 0
            vict_drug_count = 0
            vict_lgtbi_count = 0
            
            first = True
            for i in range(len(partial_df['distr'])):
                if(partial_df['week'][i] == dates[j]):
                    if(first):
                        dep.append(partial_df['dep'][i])
                        distr.append(partial_df['distr'][i])
                        fecha.append(dates[j])
                        prov.append(partial_df['prov'][i])
                        week.append(partial_df['week'][i])
                        first = False
                        
                    physical_v_count += partial_df['physical_v'][i]
                    sexual_v_count += partial_df['sexual_v'][i]
                    economical_v_count += partial_df['economical_v'][i]
                    psychological_v_count += partial_df['psychological_v'][i]
                    first_time_count += partial_df['first_time'][i]
                    inf_count += partial_df['inf'][i]
                    nin_count += partial_df['nin'][i]
                    ado_count += partial_df['ado'][i]
                    ado_t_count += partial_df['ado_t'][i]
                    jov_count += partial_df['jov'][i]
                    adu_count += partial_df['adu'][i]
                    m60_count += partial_df['m60'][i]
                    family_count += partial_df['family'][i]
                    love_count += partial_df['love'][i]
                    no_rel_count += partial_df['no_rel'][i]
                    aggr_alcohol_count += partial_df['aggr_alcohol'][i]
                    aggr_drug_count += partial_df['aggr_drug'][i]
                    vict_disab_count += partial_df['vict_disab'][i]
                    vict_alco_count += partial_df['vict_alco'][i]
                    vict_drug_count += partial_df['vict_drug'][i]
                    vict_lgtbi_count += partial_df['vict_lgtbi'][i]

            physical_v.append(physical_v_count)
            sexual_v.append(sexual_v_count)
            economical_v.append(economical_v_count)
            psychological_v.append(psychological_v_count)
            first_time.append(first_time_count)
            inf.append(inf_count)
            nin.append(nin_count)
            ado.append(ado_count)
            ado_t.append(ado_t_count)
            jov.append(jov_count)
            adu.append(adu_count)
            m60.append(m60_count)
            family.append(family_count)
            love.append(love_count)
            no_rel.append(no_rel_count)
            aggr_alcohol.append(aggr_alcohol_count)
            aggr_drug.append(aggr_drug_count)
            vict_disab.append(vict_disab_count)
            vict_alco.append(vict_alco_count)
            vict_drug.append(vict_drug_count)
            vict_lgtbi.append(vict_lgtbi_count)
            
    #armamos df y retornamos
    data = {'fecha':fecha,'distr':distr, 'prov' : prov, 'dep':dep, 'week':week, 'physical_v':physical_v,
            'sexual_v':sexual_v, 'economical_v':economical_v, 'psychological_v':psychological_v, 'first_time':first_time,
            'inf':inf, 'nin':nin, 'ado':ado,'ado_t':ado_t,'jov':jov,'adu':adu,'m60':m60,'family':family,'love':love,
            'no_rel':no_rel,'aggr_alcohol':aggr_alcohol,'aggr_drug':aggr_drug,'vict_disab':vict_disab,'vict_alco':vict_alco,
            'vict_drug':vict_drug,'vict_lgtbi':vict_lgtbi}
    response_df = pd.DataFrame(data)
    print("acabado")
    return response_df
```
